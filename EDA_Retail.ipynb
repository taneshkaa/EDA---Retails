# **Exploratory Data Analysis - Retail**
Exploratory Data Analysis on Dataset Retails (local system file - Retail.csv). EDA is performed in order to highlight the weak areas, where the organization can work in order to generate quality revenue & profit.

```python
#Importing Packages
import pandas as pd
import numpy as np
import io 
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
import matplotlib 
from matplotlib.pyplot import figure

matplotlib.rcParams['figure.figsize'] = (10,6)
```

```python
#Uploading .csv File from Local Device
from google.colab import files
uploaded = files.upload()
```

```python
# Importing Data
df = pd.read_csv(io.BytesIO(uploaded['Retail.csv']))
```

```python
# Checking for null values in the dataset
df.info()
```

```python
# Checking for missing data (if any)
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Describing the Data 
df.describe()
```

```python
# Finding unique Data for each Attribute
df.nunique()
```

```python
# Checking for duplicate values (if any)
df.duplicated().sum()
```

```python
# Droping the duplicates (if any)
df.drop_duplicates()
df.head(50)
```

```python
# Visualizing the Data using the pair plot (random pairing)
sns.pairplot(df, vars=["Sales", "Profit"])
```

```python
# Evaluating the correlation matrix
correlation_matrix = df.corr(method ='pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation matrix",fontsize=20)
plt.show()
```

```python
# Ploting relationships between categorical data like segment and profit.
sns.catplot(x="Segment",y="Profit",data=df,kind='bar',ci=97)
```

```python
# Ploting relationships between categorical data like ship mode and profit.
sns.catplot(x="Ship Mode",y="Profit",data=df,kind='bar',ci=97)
```

```python
# Ploting relationships between categorical data like state and profit.
sns.catplot(x="State",y="Profit",data=df,kind='bar',ci=97)
```

```python
# Ploting relationships between categorical data like city and profit.
sns.catplot(x="City",y="Profit",data=df,kind='bar',ci=97)
```

```python
# Ploting relationships between categorical data like postal code and profit.
sns.catplot(x="Postal Code",y="Profit",data=df,kind='bar',ci=97)
```

```python
# Profit by State
order = df.groupby(["State"])["Profit"].mean().sort_values().index
rp = sns.catplot(y="State",x="Profit",data=df,kind='bar',ci=None,height=15,order=order)
rp.fig.subplots_adjust(top=0.9)
rp.fig.suptitle('Profit by State',fontsize=30)
```

```python
# Profit by Category
rp = sns.catplot(x="Category",y="Profit",data=df,kind="bar",ci=None)
rp.fig.subplots_adjust(top=0.9)
rp.fig.suptitle('Profit by Category',fontsize=30)
```

```python
# Profit by City
rp = sns.catplot(x="City",y="Profit",data=df,kind="bar",ci=None)
rp.fig.subplots_adjust(top=0.8)
rp.fig.suptitle('Profit by Category',fontsize=30)
```

```python
# Profit by Region
sns.catplot(x="Category",y="Profit",data=df,kind='bar',ci=None,col="Region")
```

```python
# computing top categories in terms of sales from first 20 observations
top_subcategory_s = df.groupby("Category").Sales.sum().nlargest(n = 20)
# computing top categories in terms of profit from first 20 observations
top_subcategory_p = df.groupby("Category").Profit.sum().nlargest(n = 20)

# plotting to see it visually
plt.style.use('seaborn')
top_subcategory_s.plot(kind = 'bar',figsize = (13,8),fontsize = 14, color = 'brown')
top_subcategory_p.plot(kind = 'bar',figsize = (13,8),fontsize = 14)
plt.xlabel('Category',fontsize = 15)
plt.ylabel('Total Sales/Profits',fontsize = 15)
plt.title("Top Category Sales vs Profit",fontsize = 20)
plt.grid()
plt.show()
```

```python
# Count plot
plt.subplots(figsize=(12,15))
order=df.groupby(['State'])["State"].count().sort_values().index
sns.countplot(y='State',data=df,order=order)
```
